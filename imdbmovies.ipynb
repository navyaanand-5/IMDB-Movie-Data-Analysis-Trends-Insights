To handle the null values by Fills Missing Values with Its Mean

```python
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].mean())
df['Metascore'] = df['Metascore'].fillna(df['Metascore'].mean())
df.isna().sum()
```

Converting string data into numeric data

```python
# Title
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
out = ohe.fit_transform(df[['Title']])
out
```

```python
import pandas as pd
df1 = pd.DataFrame(out,columns=ohe.get_feature_names_out())
df1.head()
```

```python
data = pd.concat([df,df1],axis=1)
```

```python
data.drop('Title',axis=1,inplace=True)
data.head()
```

#  DATA VISUALIZATION



1. To check in which year there was the highest voting.

```python
df.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Votes',data=df,color='y')
plt.title('Votes by Year')
plt.xlabel('Year')
plt.ylabel('Votes')
plt.show()
```

2. In which year there was highest revenue?  So the below graph shows the distribution of movie revenue by year from 2006 to 2016.



```python
df.groupby('Year')['Revenue (Millions)'].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x='Year', y='Revenue (Millions)', data=df,color='c')
plt.xticks(rotation=45)
plt.title('Revenue Distribution by Year')
plt.show()
```

3. Visualizing the proportion of high-rated (7) and low-rated (<7) movies in the dataset.

```python
high_rated = df[df['Rating'] >= 7].shape[0]
low_rated = df[df['Rating'] < 7].shape[0]

plt.figure(figsize=(5, 5))
plt.pie([high_rated, low_rated], labels=["High Rated (7)", "Low Rated (<7)"],autopct='%1.1f%%', colors=['blue', 'magenta'], startangle=140)
plt.title("High vs. Low Rated Movies")
plt.show()
```

4.  To represent the number of movies released per year from 2006 to 2016.

```python
df1 = df['Year'].value_counts()
```

```python
df1.plot(linestyle='solid', marker='*', color='c')
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.title("Movies Released Per Year (2006-2016)")
plt.grid(True)
plt.show()
```

5. Displaying top 5 lengthy movie title and runtime

```python
df['Title_Length'] = df['Title'].apply(len)
```

```python
top_len = df.nlargest(5,'Runtime (Minutes)')[['Title','Runtime (Minutes)']]
```

```python
sns.barplot(x='Runtime (Minutes)',y='Title',data=top_len,width=0.3,palette='viridis')
plt.title('Title and Runtime')
plt.show()
```

6. showing the distribution of movie genres in the dataset. Each bar represents the count of movies belonging to a particular genre.

```python
df2 = df['Genre'].str.split(',').explode().value_counts()

plt.figure(figsize=(10, 5))
plt.bar(df2.index, df2.values, color='skyblue')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Distribution of Movie Genres')
plt.xticks(rotation=45)
plt.show()
```

7. Top 10 highest revenue movie titles.

```python
top_10 = df.nlargest(10,'Revenue (Millions)')[['Title','Revenue (Millions)']]
```

```python
plt.figure(figsize=(5, 4))
sns.barplot(x='Revenue (Millions)',y='Title',data=top_10,color='c')
plt.title('Top 10 Movies')
plt.show()
```

8. Identify relationships between numerical variables in the dataset.

```python
numeric_df = df[["Year", "Runtime (Minutes)", "Rating", "Votes", "Revenue (Millions)", "Metascore"]]
corr_matrix = numeric_df.corr()

plt.figure(figsize=(7,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Movie Dataset")
plt.show()
```

9. Does rating affect the revenue?

```python
sns.scatterplot(x='Rating',y='Revenue (Millions)',data=df,color='r')
plt.title('Rating vs. Revenue')
plt.show()
```
